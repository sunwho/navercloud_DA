```python
def snap(d1):
    od_list1 = []
    pt_list1 = []
    od_list2 = []
    pt_list2 = []
    od_list3 = []
    pt_list3 = []
    od_list4 = []
    pt_list4 = []
    list2 = [od_list1, pt_list1, od_list2, pt_list2, od_list3, pt_list3, od_list4, pt_list4]
    
    for i in range(0, len(d1)):
        d = d1[i].copy()  # 원본 데이터프레임을 복사하여 작업
        d['미터링일시'] = d['미터링일시'].astype(str)  # 열을 문자열로 변환
        
        for j in range(1, 32):
            df = d[d['미터링일시'].str[-2:] == str(j).zfill(2)]
            list2[i].append(df)
    
    return list2

all_cor = snap(cor_list1)
all_drv = snap(drv_list1)
all_app = snap(app_list1)
```

```python
def new_domain (d1, d2) :
    d2_key = d2['회원번호'].drop_duplicates()
    d1_cleaned = d1.merge(d2_key, on='회원번호', how='left', indicator=True)
    d1_cleaned = d1_cleaned[d1_cleaned['_merge'] == 'left_only']
    d1_cleaned = d1_cleaned.drop(columns=['_merge'])
    d1_cleaned = d1_cleaned.reset_index(drop=True)
    return d1_cleaned
```
