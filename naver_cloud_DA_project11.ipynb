```python
def format_minutes(dt_str):
    date_part, time_part = dt_str.split(' ')
    hour, minute = time_part.split(':')
    minute = minute.zfill(2)
    return f"{date_part} {hour}:{minute}"
```

```python
def date_refine(d1) :
    d1['registdate'] = pd.to_datetime(d1['registdate'], format='%Y-%m-%d %H:%M:%S.%f')
    d1['registdate'] = d1['registdate'].dt.strftime('%Y.%m.%d %H:%M').str.replace(r'\b0+(?!\b)', '', regex=True)
    d1['registdate'] = d1['registdate'].apply(format_minutes)
    return (d1)
```
