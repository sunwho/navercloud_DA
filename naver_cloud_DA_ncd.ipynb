```python
def preprocess (df) :
    df['member_id'] = df['회원번호']
    df['calc_ymd'] = df['미터링일시']
    df['seller_id'] = df['파트너사회원번호']
    df['seller_id'] = df['seller_id'].fillna(0)
    df['seller_id'] = df['seller_id'].astype(int)
    min_dates = df.groupby(['member_id', 'plan'])['calc_ymd'].min().reset_index()
    min_dates = min_dates.rename(columns={'calc_ymd': 'regist_ymd'})
    df = df.merge(min_dates, on=['member_id', 'plan'], how='left')
    max_dates = df.groupby(['member_id', 'plan'])['calc_ymd'].max().reset_index()
    max_dates = max_dates.rename(columns={'calc_ymd': 'delete_ymd'})
    df = df.merge(max_dates, on=['member_id', 'plan'], how='left')

    from datetime import datetime
    today_int = int(datetime.today().strftime('%Y%m%d'))
    
    df.loc[:, 'leaveYn'] = np.where(df['delete_ymd'] == today_int-1, 'N', 'Y')
    df['delete_ymd'] = df['delete_ymd'].astype(float)
    df.loc[:, 'delete_ymd'] = np.where(df['delete_ymd'] == today_int-1, np.nan, df['delete_ymd'])
    df['delete_ymd'] = df['delete_ymd'].apply(lambda x: int(x) if pd.notna(x) else pd.NA).astype('Int64')
    df = df.sort_values(by=['member_id','calc_ymd','사용량'], ascending = [True, True, False])
    df = df.drop_duplicates(subset = ['member_id','calc_ymd'], keep = 'first')
    return df
```
