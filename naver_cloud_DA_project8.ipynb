```python
def format_minutes(dt_str):
    date_part, time_part = dt_str.split(' ')
    hour, minute = time_part.split(':')
    minute = minute.zfill(2)
    return f"{date_part} {hour}:{minute}"
```

```python
def date_refine(d1) :
    d1['registdate'] = pd.to_datetime(d1['registdate'], format='%Y-%m-%d %H:%M:%S.%f')
    d1['registdate'] = d1['registdate'].dt.strftime('%Y.%m.%d %H:%M').str.replace(r'\b0+(?!\b)', '', regex=True)
    d1['registdate'] = d1['registdate'].apply(format_minutes)
    return (d1)
```

```python
def new_app_domain (d1) :
    list1 = []
    
    d1_app = d1[(d1['plan'] == 'APP_STD')|(d1['plan'] == 'ATT_STD')|(d1['plan'] == 'FIN_STD')|(d1['plan'] == 'PAY_STD')]
    app_grouped = d1_app.groupby(['domainId', 'odPartnerstatus'])['plan'].apply(lambda x: frozenset(x)).reset_index()
    
    possible_combinations = {
        frozenset(['APP_STD']): 'Only APP_STD',
        frozenset(['APP_STD', 'ATT_STD']): 'APP_STD + ATT_STD',
        frozenset(['APP_STD', 'FIN_STD']): 'APP_STD + FIN_STD',
        frozenset(['APP_STD', 'PAY_STD']): 'APP_STD + PAY_STD',
        frozenset(['APP_STD', 'ATT_STD', 'FIN_STD']): 'APP_STD + ATT_STD + FIN_STD',
        frozenset(['APP_STD', 'ATT_STD', 'PAY_STD']): 'APP_STD + ATT_STD + PAY_STD',
        frozenset(['APP_STD', 'FIN_STD', 'PAY_STD']): 'APP_STD + FIN_STD + PAY_STD',
        frozenset(['APP_STD', 'ATT_STD', 'FIN_STD', 'PAY_STD']): 'APP_STD + ATT_STD + FIN_STD + PAY_STD',
    }
    
    app_grouped['plan_combination'] = app_grouped['plan'].map(possible_combinations)
    app_grouped['plan_combination_with_status'] = app_grouped['plan_combination'] + " (" + app_grouped['odPartnerstatus'] + ")"
    result = app_grouped.groupby('plan_combination_with_status').size().reset_index(name='count')
    
    ordered_combinations = []
    for comb in possible_combinations.values():
        ordered_combinations.append(comb + " (OD)")
        ordered_combinations.append(comb + " (PARTNER)")
    
    result = result.set_index('plan_combination_with_status').reindex(ordered_combinations, fill_value=0)
    list1 = result['count'].tolist()
    return list1
```

```python
def lastday_snapshot_user (d1) :
    d1_od_cor_lt = d1[(d1['plan'] == 'COR_LT') & (d1['odPartnerstatus'] == 'OD')]
    d1_pt_cor_lt = d1[(d1['plan'] == 'COR_LT') & (d1['odPartnerstatus'] == 'PARTNER')]
    d1_od_cor_std = d1[(d1['plan'] == 'COR_STD') & (d1['odPartnerstatus'] == 'OD')]
    d1_pt_cor_std = d1[(d1['plan'] == 'COR_STD') & (d1['odPartnerstatus'] == 'PARTNER')]
    d1_od_cor_std_p = d1[(d1['plan'] == 'COR_STD_P') & (d1['odPartnerstatus'] == 'OD')]
    d1_pt_cor_std_p = d1[(d1['plan'] == 'COR_STD_P') & (d1['odPartnerstatus'] == 'PARTNER')]
    d1_od_drv_lt = d1[(d1['plan'] == 'DRV_LT') & (d1['odPartnerstatus'] == 'OD')]
    d1_pt_drv_lt = d1[(d1['plan'] == 'DRV_LT') & (d1['odPartnerstatus'] == 'PARTNER')]
    d1_od_drv_std = d1[(d1['plan'] == 'DRV_STD') & (d1['odPartnerstatus'] == 'OD')]
    d1_pt_drv_std = d1[(d1['plan'] == 'DRV_STD') & (d1['odPartnerstatus'] == 'PARTNER')]
    d1_od_drv_std_p = d1[(d1['plan'] == 'DRV_STD_P') & (d1['odPartnerstatus'] == 'OD')]
    d1_pt_drv_std_p = d1[(d1['plan'] == 'DRV_STD_P') & (d1['odPartnerstatus'] == 'PARTNER')]
    
    d1_app = d1[(d1['plan'] == 'APP_STD')|(d1['plan'] == 'ATT_STD')|(d1['plan'] == 'FIN_STD')|(d1['plan'] == 'PAY_STD')]
    
    app_grouped = d1_app.groupby(['domainId', 'odPartnerstatus']).agg({'plan': lambda x: frozenset(x), 'userCount': 'sum'}).reset_index()

    possible_combinations = {
        frozenset(['APP_STD']): 'Only APP_STD',
        frozenset(['APP_STD', 'ATT_STD']): 'APP_STD + ATT_STD',
        frozenset(['APP_STD', 'FIN_STD']): 'APP_STD + FIN_STD',
        frozenset(['APP_STD', 'PAY_STD']): 'APP_STD + PAY_STD',
        frozenset(['APP_STD', 'ATT_STD', 'FIN_STD']): 'APP_STD + ATT_STD + FIN_STD',
        frozenset(['APP_STD', 'ATT_STD', 'PAY_STD']): 'APP_STD + ATT_STD + PAY_STD',
        frozenset(['APP_STD', 'FIN_STD', 'PAY_STD']): 'APP_STD + FIN_STD + PAY_STD',
        frozenset(['APP_STD', 'ATT_STD', 'FIN_STD', 'PAY_STD']): 'APP_STD + ATT_STD + FIN_STD + PAY_STD',
    }
    
    app_grouped['plan_combination'] = app_grouped['plan'].map(possible_combinations)

    result = app_grouped.groupby(['plan_combination', 'odPartnerstatus'])['userCount'].sum().unstack(fill_value=0)
    
    result['TOTAL'] = result.sum(axis=1)
    
    result = result.reindex(possible_combinations.values(), fill_value=0)
    
    item_usage = {
        'COR_LT' : [d1_od_cor_lt['userCount'].sum(),d1_pt_cor_lt['userCount'].sum(),(d1_od_cor_lt['userCount'].sum()+d1_pt_cor_lt['userCount'].sum())],
        'COR_STD' : [d1_od_cor_std['userCount'].sum(),d1_pt_cor_std['userCount'].sum(),(d1_od_cor_std['userCount'].sum()+d1_pt_cor_std['userCount'].sum())],
        'COR_STD_P' : [d1_od_cor_std_p['userCount'].sum(),d1_pt_cor_std_p['userCount'].sum(),(d1_od_cor_std_p['userCount'].sum()+d1_pt_cor_std_p['userCount'].sum())],
        
        'COR TOTAL' : [d1_od_cor_lt['userCount'].sum()+d1_od_cor_std['userCount'].sum()+d1_od_cor_std_p['userCount'].sum(),
                   d1_pt_cor_lt['userCount'].sum()+d1_pt_cor_std['userCount'].sum()+d1_pt_cor_std_p['userCount'].sum(),
                   d1_od_cor_lt['userCount'].sum()+d1_od_cor_std['userCount'].sum()+d1_od_cor_std_p['userCount'].sum()+d1_pt_cor_lt['userCount'].sum()+d1_pt_cor_std['userCount'].sum()+d1_pt_cor_std_p['userCount'].sum()],
        
        'DRV_LT' : [d1_od_drv_lt['userCount'].sum(),d1_pt_drv_lt['userCount'].sum(),(d1_od_drv_lt['userCount'].sum()+d1_pt_drv_lt['userCount'].sum())],
        'DRV_STD' : [d1_od_drv_std['userCount'].sum(),d1_pt_drv_std['userCount'].sum(),(d1_od_drv_std['userCount'].sum()+d1_pt_drv_std['userCount'].sum())],
        'DRV_STD_P' : [d1_od_drv_std_p['userCount'].sum(),d1_pt_drv_std_p['userCount'].sum(),(d1_od_drv_std_p['userCount'].sum()+d1_pt_drv_std_p['userCount'].sum())],
        
        'DRV TOTAL' : [d1_od_drv_lt['userCount'].sum()+d1_od_drv_std['userCount'].sum()+d1_od_drv_std_p['userCount'].sum(),
                   d1_pt_drv_lt['userCount'].sum()+d1_pt_drv_std['userCount'].sum()+d1_pt_drv_std_p['userCount'].sum(),
                   d1_od_drv_lt['userCount'].sum()+d1_od_drv_std['userCount'].sum()+d1_od_drv_std_p['userCount'].sum()+d1_pt_drv_lt['userCount'].sum()+d1_pt_drv_std['userCount'].sum()+d1_pt_drv_std_p['userCount'].sum()]
}
    item_usage_summary = pd.DataFrame(item_usage).T
    item_usage_summary.columns = ['OD','PARTNER','TOTAL']
    summary = pd.concat([item_usage_summary,result])
    summary.to_excel('Desktop/8summary.xlsx', index=True)
    return summary
```
